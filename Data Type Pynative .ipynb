## List data type

```python
my_list=["hey" , 12 , 34.5 ,True , False]
# Display list
print(my_list)
```

```python
print(type(my_list)) # class list
```

```python
print(my_list[2]) # accessing third element of list
```

```python
print(my_list[0:3]) # slicing
```

```python
print(my_list[::-1]) # reverse slicing
```

 ## Dictionary data type
