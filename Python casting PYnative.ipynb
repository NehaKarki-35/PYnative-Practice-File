***Casting float value to an integer***

```python
value=25.3
print(type(value))
v=int(value)
print(v)
```

```python
print(value)
```

```python
value=int(25.5)
print("Integer value :" ,value)
```

```python
print(type(value))
```

***Casting Boolean value to an integer***

```python
option1=True
option2=False
print(type(option1))
print(type(option2))
```

```python
# converting bool in integer
num1=int(option1)
num2=int(option2)
print(num1,num2)
```

```python
print("Integer number:",num1,num2)
```

***Casting integer type to complex type***

```python
num=1234
print(type(num))
```

```python
num=complex(num)
print("complex num:",num)
```

***Casting float type to complex type***

```python
num1=123.34
print(type(num1))
```

```python
num1=complex(num1)
print(num1)
```

***Casting Boolean type to complex type***

```python
bool_value=True
print(type(bool_value))
```

```python
bool_value_=complex(bool_value)
print("boolean value:", bool_value_)
```

```python
print(type(bool_value_))
```

```python
num1,num2=True,False
num3=complex(num1,num2)
print("num3:",num3)
```

```python
type(num3)
```
